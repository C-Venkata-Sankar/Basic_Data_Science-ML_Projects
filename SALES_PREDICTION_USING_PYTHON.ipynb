```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/advertising.csv')
```

```python
x = df.drop(columns = 'Sales')
y = df['Sales']
```

```python
model = RandomForestClassifier(random_state=0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
bins = np.linspace(min(y_train), max(y_train), num=5)  # Adjust num as needed
y_train_categorical = np.digitize(y_train, bins)
y_test_categorical = np.digitize(y_test, bins)
```

```python
model.fit(x_train, y_train_categorical)
```

```python
y_train_pred = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train_categorical, y_train_pred)
print(training_data_accuracy)
```

```python
y_test_predict = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test_categorical, y_test_predict)
print(test_data_accuracy)
```
